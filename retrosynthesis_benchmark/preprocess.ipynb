```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Draw
import numpy as np
import seaborn as sns
```

```python
df = pd.read_pickle(r'uspto_50.pickle')
```

```python
df['reactants_mol'] = df.apply(lambda x: [Chem.CanonSmiles(smi) for smi in Chem.MolToSmiles(x['reactants_mol']).split('.')], axis=1)
df['products_mol'] = df.apply(lambda x: [Chem.CanonSmiles(smi) for smi in Chem.MolToSmiles(x['products_mol']).split('.')], axis=1)
```

```python
df['num_reacts'] = df.apply(lambda x: len(x['reactants_mol']), axis=1)
df['num_prods'] = df.apply(lambda x: len(x['products_mol']), axis=1)
```

```python
df['ratio'] = df.apply(lambda x: sum([Chem.MolFromSmiles(smi).GetNumAtoms() for smi in x['reactants_mol']])/Chem.MolFromSmiles(x['products_mol'][0]).GetNumAtoms(), axis=1)
```
